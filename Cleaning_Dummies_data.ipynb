Let's bring the dummies variables into regular variables combining into one variable on Flooring_ and District_

```python
flooring= []

# scan all features starting with Flooring_ and put in flooring set
for col in df.columns:
  if col.startswith("Flooring_"):
    flooring.append(col)

# apply flooring data into original dataset naming FlooringType
df["FlooringType"]=df[flooring].apply(lambda x: x.idxmax(), axis=1)
df["FlooringType"] = df["FlooringType"].str.replace("Flooring_", "")

# delete all the features that starts with Flooring_ in original dataset
df=df.drop(columns=flooring)
```

Let's double check.

```python
for col in df.columns:
    print(col)
```

Now we will do same work on District_ features as what we did above.

```python
District = []

# scan all the District_ features in orginal dataset and put in District set
for col in df.columns:
  if col.startswith("District_"):
    District.append(col)

# put all those features in one feature called SchoolDistrict
df["SchoolDistrict"]=df[District].apply(lambda x: x.idxmax(), axis=1)
df["SchoolDistrict"] = df["SchoolDistrict"].str.replace("Flooring_", "")

# delete the features starts with District_
df=df.drop(columns=District)
```
